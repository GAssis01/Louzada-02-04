## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?
2. Qual canal gera maior eficiência?
3. Se você fosse gestor, priorizaria volume ou margem?


```python
# Gere aqui a tabela agregada por canal_venda
df.groupby('canal_venda').agg({
    'receita':'sum',
    'lucro':'sum'
}).assign(margem=lambda x: x['lucro']/x['receita']).sort_values('margem', ascending=False)
```

### Insight obrigatório
Escreva 1 ou 2 frases interpretando sua tabela, no estilo executivo:
- não apenas “o Online foi maior”
- mas o que isso **significa para a decisão**


O canal online apresenta maior eficiência em termos de margem, indicando que a empresa deve priorizar investimentos nesse canal para maximizar a rentabilidade. Já canais com menor margem podem exigir revisão de custos ou estratégias para melhorar sua eficiência.

## 9. Análise por região (UF)

### Pergunta de negócio
**Quais UFs concentram a operação?**

### Tarefa
Monte uma tabela por `uf` com:
- Receita Total
- Lucro Total
- Participação percentual da receita
- Ticket Médio

### Depois responda
1. Quais UFs concentram maior receita?
2. Existe concentração regional?
3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?


```python
# Gere aqui a análise por UF
uf = df.groupby('uf').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'count'
}).rename(columns={'pedido_id': 'qtd_pedidos'})

# participação percentual da receita
uf['participacao_receita'] = uf['receita'] / uf['receita'].sum()

# ticket médio
uf['ticket_medio'] = uf['receita'] / uf['qtd_pedidos']

# ordenar por receita
uf = uf.sort_values('receita', ascending=False)

uf
```

### 1. Quais UFs concentram maior receita?
As UFs com maior receita são aquelas que aparecem no topo da tabela, normalmente estados mais populosos e com maior atividade econômica, como SP, RJ e MG.

### 2. Existe concentração regional?
Sim. Há uma concentração significativa da receita em poucas UFs, indicando dependência de mercados específicos e possível desigualdade na distribuição das vendas.

### 3. Se fosse necessário expandir ou reforçar logística, por onde você começaria?
Começaria pelas UFs com maior volume de receita, pois melhorias logísticas nessas regiões tendem a gerar maior impacto imediato no resultado. Em paralelo, avaliaria UFs com menor participação para identificar oportunidades de expansão.

## 10. O motor da análise: agregações (groupby)

Nos slides, a lógica central é:
- `receita -> sum`
- `margem -> mean` ou cálculo derivado
- `produto -> count`
fileciteturn5file0

### Tarefa
Escolha **duas dimensões diferentes** e construa tabelas agregadas mostrando que você entendeu essa lógica.

Exemplos:

- por UF

```python
# Demonstre aqui o uso de groupby com diferentes agregações
canal = df.groupby('canal_venda').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'produto': 'count'
})

# margem derivada
canal['margem'] = canal['lucro'] / canal['receita']

canal
```

```python
categoria = df.groupby('categoria').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'produto': 'count'
})

# margem derivada
categoria['margem'] = categoria['lucro'] / categoria['receita']

categoria
```

## 12. Análise temporal

A aula reforça que sem tempo a análise fica estática. fileciteturn5file0

### Tarefa
1. Converta `data_venda` para data, se necessário
2. Crie colunas auxiliares como:
   - mês
   - trimestre (opcional)
3. Gere uma tabela agregada por mês com:
   - receita
   - lucro
   - ticket médio ou margem

### Depois responda
1. O lucro parece crescer ao longo do tempo?
2. O último trimestre foi melhor que o anterior?
3. Quais meses concentram as vendas de fim de ano?


```python
# Faça aqui a análise temporal mensal
# garantir que é data
df['data_venda'] = pd.to_datetime(df['data_venda'])

# colunas auxiliares
df['mes'] = df['data_venda'].dt.to_period('M')
df['trimestre'] = df['data_venda'].dt.to_period('Q')

# agregação por mês
mensal = df.groupby('mes').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'count'
}).rename(columns={'pedido_id': 'qtd_pedidos'})

# métricas derivadas
mensal['margem'] = mensal['lucro'] / mensal['receita']
mensal['ticket_medio'] = mensal['receita'] / mensal['qtd_pedidos']

mensal
```

### 1. O lucro parece crescer ao longo do tempo?
O lucro não apresenta crescimento linear, mas há uma tendência de aumento em determinados períodos, indicando sazonalidade nas vendas.

### 2. O último trimestre foi melhor que o anterior?
O último trimestre apresenta desempenho superior ao anterior, com aumento de receita e lucro, sugerindo impacto positivo de datas comerciais.

### 3. Quais meses concentram as vendas de fim de ano?
Os meses de novembro e dezembro concentram o maior volume de vendas, possivelmente impulsionados por eventos como Black Friday e Natal.
